# Bulk Download

## Target

**정현: 2022년 HS6 전품목 구다사이**

```python
# months in 2022

months = [f"2022{month:02d}" for month in range(1, 13)]
months
```

```python
# all country

reporters_filtered = reporters[~(reporters['entryExpiredDate'] < '2022-01-01')]
reporters_filtered = reporters_filtered.loc[:, ['reporterCode', 'reporterDesc', 'reporterCodeIsoAlpha3']]
reporters_filtered = reporters_filtered.reset_index(drop=True)
reporters_filtered
```

```python
# hs6 only (추후 필터링용)

hscodes_filtered = hscodes[hscodes['aggrlevel'] == 6]
hscodes_filtered = hscodes_filtered.loc[:, ['id', 'text']]
hscodes_filtered = hscodes_filtered.reset_index(drop=True)
hscodes_filtered
```

## Bulk Availiavility

```python
# check

targets = pd.DataFrame()

for month in months:

    for reporter in tqdm(reporters_filtered['reporterCode'], desc='Checking Bulk Availability'):

        temp = cc.getFinalDataBulkAvailability(
            subscription_key, typeCode='C', freqCode='M', clCode='HS',
            period=month, reporterCode=reporter, # parameters
            publishedDateFrom=None, publishedDateTo=None
            )

        targets = pd.concat([targets, temp], axis=0)

    targets.to_csv(os.path.join(main_dir, f'bulk_availability_temp.csv'), index=False)

os.remove(os.path.join(main_dir, f'bulk_availability_temp.csv'))
targets.to_csv(os.path.join(main_dir, f'bulk_availability.csv'), index=False)
```

```python
# target list

targets = pd.read_csv(os.path.join(main_dir, 'bulk_availability.csv'))
targets
```

```python
# counting

temp1 = targets[targets['classificationCode'] == 'H6'] # HS6 기준 작성

counts = temp1['reporterCode'].value_counts()
temp2 = temp1[temp1['reporterCode'].isin(counts[counts == 12].index)] # 12개월 full records

print(' '.join([f"Total: {len(months) * len(reporters_filtered)} /",
                f"Bulk Available: {len(targets)} /",
                f"Full Records: {len(temp1)} /",
                f"Full Records & HS6 Used: {len(temp2)}"])) # 71*12 = 852
```

## Download

- `freqCode` = 'M'
- `clCode` = 'HS'
- `period` = 202201 ~ 202212, 12개월
- `reporterCode` = 4 ~ 716, 222개국 ➡ **최종 71개국 (Full Records, HS6)**

```python
# final list

targets_filtered = temp2.loc[:, ['period', 'reporterCode']]
targets_filtered = targets_filtered.reset_index(drop=True)
targets_filtered = targets_filtered.merge(reporters_filtered, on='reporterCode', how='left')
targets_filtered['temp'] = targets_filtered['reporterCode'].apply(lambda x: f"{x:03}")

for i in range(len(targets_filtered)):
    targets_filtered.loc[i, 'file_name'] = f"COMTRADE-FINAL-CM{targets_filtered['temp'][i]}{targets_filtered['period'][i]}H6" # 다운로드 체크용

targets_filtered.drop(columns=['temp'], inplace=True)
targets_filtered.to_csv(os.path.join(main_dir, 'bulk_download_target.csv'), index=False)
```

```python
targets_filtered = pd.read_csv(os.path.join(main_dir, 'bulk_download_target.csv'))
targets_filtered
```

```python
for i in tqdm(range(len(targets_filtered)), desc='Downloading Files'):

    cc.bulkDownloadFinalFile(
        subscription_key, down_dir, typeCode='C', freqCode='M',clCode='HS',
        period=targets_filtered['period'][i], reporterCode=targets_filtered['reporterCode'][i], # parameters
        decompress=True
        )
```

# Check

```python
downloaded_files = os.listdir(down_dir)
checks = targets_filtered['file_name'].apply(lambda x: any(f.startswith(x) for f in downloaded_files))

print(f"Downloaded: {len(downloaded_files)} / Leaks: {len(targets_filtered) - sum(checks)}")
```
